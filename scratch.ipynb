```python
sigma = [1,1]                            #n_dimensions
bounds = [[-5,-2],[5,5]] #[[-2,0],[2,3]] #   #2 x n_dimensions
nbins = (11,15) #(5,4) #                    #n_dimensions

mgrid_args = tuple(slice(bounds[0][d], bounds[1][d], nbins[d]*1j) for d in range(len(bounds)))
mg = np.mgrid[mgrid_args]
coords = np.array([[-1,0],[3,4],[0,-4]]) #n_walkers x n_dimensions

#sigma matrix
s0 = np.tile(sigma, (len(coords),*nbins,1)).transpose(0,3,1,2)
print(s0.shape)
# print(s0)

#walker coordinate matrix
x0 = np.tile(coords, nbins+(1,1)).transpose(2,3,0,1) #= np.full((10,10), 5)
print(x0.shape)
# print(x0)

#bincenter matrix
mgrid_tiled = np.tile(mg, (len(coords),1,1,1))
print(mgrid_tiled.shape)
# print(mgrid_tiled)

expcomps = np.exp(-((x0-mgrid_tiled)/(2*s0))**2)
print(expcomps)
print(expcomps.shape)
exps = np.product(expcomps, axis = 1)
print(exps)
print(exps.shape)
for e in exps:
    plt.imshow(e, vmax=1.1, extent = np.array(bounds).transpose().flatten(), aspect = (bounds[1][0]-bounds[0][0])/(bounds[1][1]-bounds[0][1]))
    #plt.axis("equal")
    plt.show()

exptot = np.sum(exps, axis = 0)
print(np.max(exptot))
plt.imshow(exptot, vmax=1.1, extent = np.array(bounds).transpose().flatten(), aspect = (bounds[1][0]-bounds[0][0])/(bounds[1][1]-bounds[0][1]))
plt.show()
```
